```python
import pandas as pd
```

Cargar datos

```python
# Importamos el dataset de los datos de los pokemon
ds = pd.read_csv(r'dataset_final.csv')
```

```python
'''
1. ¿Cuántos pokémon hay en el dataset?
'''
ds['#'].count()
```

```python
'''
2. ¿Cuántos pokémon hay de tipo Poison? (Type 1)
'''
ds.groupby('Type 1').get_group('Poison')['#'].count()
```

```python
'''
3. ¿Cuántos tipos diferentes de pokémon hay? (Type 1)
'''
len(ds.groupby('Type 1').groups)
```

```python
'''
2. ¿Cuántos pokémon hay de cada tipo? (Type 1)
'''
print("La cantidad de pokemons por tipo es:")
print(ds.groupby('Type 1')['#'].count())
'''
2.1 ¿cual es el tipo de pokémon con más pokémon? (Type 1)
'''
sorted = ds.groupby('Type 1')['#'].count().sort_values(ascending = False)
maxvalue = sorted[0]
maxvaluename = sorted.index[0]
print(f"\nEl tipo con más pokemons es {maxvaluename} con {maxvalue} criaturitas")
```

```python
'''
3. ¿Cuál es el pokémon más rápido?
'''
pokemon = ds.sort_values(by=['Speed'] , ascending = False).iloc[0]

print(f"El pokemon más veloz es {pokemon['Name']}")
```

```python
'''
4. ¿Cuántos pokemon tiene una defensa superior a 100?
'''

'''
4.1 ¿Cuántos pokemon tiene una defensa superior a 100?
'''
condition = ds['Defense']>100
defense_ds = ds[condition]
count_def100 = defense_ds['#'].count()
names_def100 = ", ".join(defense_ds['Name'].values)

print(f"Hay {count_def100} pokemon con una defensa superior a 100 y son: {names_def100}")
```

```python
'''
5. ¿Cuántos pokemon tiene una defensa superior a 100 y una velocidad superior a 100?
'''

condition = (ds['Defense']>100) & (ds['Speed']>100)
ds[condition]['#'].count()
```

```python
'''
6. Ordena el dataset por el tipo 1 y por el tipo 2
'''
ds.sort_values(['Type 1', 'Type 2'])
```

```python
'''
7. Crea un nuevo dataset con los pokémon de tipo Water y Fire como primer tipo
'''
water_condition = ds['Type 1'] == "Water"
fire_condition = ds['Type 1'] == "Fire"
water_fire_ds = ds[water_condition | fire_condition]
water_fire_ds
```

```python
'''
8. Crea un nuevo dataset con los pokemon Legendary
'''
legendary_condition = ds['Legendary']
legendaries = ds[legendary_condition]
legendaries
```

```python
'''
9. Calcula el promedio de stats de los pokemon Legendary (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed) y los no Legendary
'''

non_legendary_condition = ds['Legendary'] == False
non_legendaries = ds[non_legendary_condition]

def average(dataset):
    avg_count = dataset['#'].count()
    avg = dataset[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sum().apply(lambda value: value/avg_count)
    return avg

legen_average = average(legendaries)
non_legen_average = average(non_legendaries)
print("Estos son los promedios de los stats de los legendarios \n")
print(legen_average)
print("\nEstos son los promedios de los stats de los no legendarios \n")
print(non_legen_average)
```

```python
'''
10. Crea un nuevo dataframe con el resultado del anterior ejercicio comparando ambos promedios

ejemplo:
                HP  Attack  Defense  Sp. Atk  Sp. Def  Speed
Legendary       99      90       89       91       94     90
No Legendary    80      95       75       12       43     87

'''

pd.DataFrame(data=[legen_average, non_legen_average], index = ['Legendary', 'No Legendary'])
```

```python
'''
11. Añade una nueva colunma ['Doble tipo'] al dataframe inicial que inndique si el pokemo tiene dos tipos o no
'''
dtype_condition = (ds['Type 1'].notna()) & (ds['Type 2'].notna())
ds['Doble Tipo'] = dtype_condition
```

```python
'''
12. Muestra las columas Name, Type 1, Type 2 de los pokémon que tienen dos tipos y ordenalos por Type 1 , Type 2 y Name
'''

ds[dtype_condition][['Name', 'Type 1', 'Type 2']].sort_values(['Type 1', 'Type 2' , 'Name'])
```
